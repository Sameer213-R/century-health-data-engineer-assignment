## checking validation

```python
master = spark.table("gold.master_health")

print("Total Rows:", master.count())

master.selectExpr(
    "count(distinct patient) as unique_patients",
    "count(distinct encounter_id) as unique_encounters"
).show()

# Null Checks
master.selectExpr(
    "sum(case when patient is null then 1 else 0 end) as null_patients",
    "sum(case when encounter_id is null then 1 else 0 end) as null_encounters"
).show()

# Duplicate Encounter Rows
master.groupBy("encounter_id").count().orderBy("count", ascending=False).show(10)

# Sample Data
master.show(10, truncate=False)
```
