```python
%sql 
use catalog centuryhealthdatabase;
use schema bronze
```

```python
patients = load_csv("patients.csv")
symptoms = load_csv("symptoms.csv")
medications = load_csv("medications.csv")
conditions = load_csv("conditions.csv")
encounters = load_parquet("encounters.parquet")

patients.write.mode("overwrite").format("delta").saveAsTable("bronze.patients")
symptoms.write.mode("overwrite").format("delta").saveAsTable("bronze.symptoms")
medications.write.mode("overwrite").format("delta").saveAsTable("bronze.medications")
conditions.write.mode("overwrite").format("delta").saveAsTable("bronze.conditions")
encounters.write.mode("overwrite").format("delta").saveAsTable("bronze.encounters")

print("Bronze Layer Complete")
```
