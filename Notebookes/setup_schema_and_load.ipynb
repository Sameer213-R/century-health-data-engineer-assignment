## file path variable


```python
RAW_PATH = "/Volumes/centuryhealthdatabase/centuryfileholder/rawfile/"
```

## load function

```python
def load_csv(name):
    return spark.read.option("header",True).option("inferSchema",True).csv(RAW_PATH + name)

def load_parquet(name):
    return spark.read.parquet(RAW_PATH + name)

def load_excel(name):
    return spark.read.format("com.crealytics.spark.excel") \
        .option("header","true") \
        .option("inferSchema","true") \
        .load(RAW_PATH + name)

print("Schemas Created Successfully")
```
